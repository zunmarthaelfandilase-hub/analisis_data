```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Pengaturan tampilan
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12

print('Semua package berhasil diimport!')
```

```python
# LINK
link = 'https://docs.google.com/spreadsheets/d/1giZNTO-G7Iiytorc2ZuhbMZhvMa9KWX8/export?format=csv'

# Baca data
data = pd.read_csv(link)

# Tampilkan beberapa baris pertama dari data untuk verifikasi
print("\n\n beberapa baris pertama dari data untuk verifikasi :")
print(data.head(6))
```

```python
print("\n\n 6 Data Pertama :")
print(data.head(6))
```

```python
print("\n\n 6 Data Terakhir :")
print(data.tail(6))
```

```python
print('=== STRUKTUR DATA ===')
print(f'Jumlah Baris   : {data.shape[0]}')
print(f'Jumlah Kolom   : {data.shape[1]}')
print()
print('--- Tipe Data Setiap Kolom ---')
print(data.dtypes)
print()
print('--- Nilai Null per Kolom ---')
print(data.isnull().sum())
```

```python
print("\n\n Tampilan Ringkasan Statistik Awal")
print(data.describe())
```

```python
baris, kolom = data.shape
print("\n\n Jumlah Baris dan Kolom")
print(f"jumlah baris : {data.shape[0]}")
print(f"jumlah kolom : {data.shape[1]}")
```

```python
print('=== DISTRIBUSI KATEGORI PELANGGAN (Segment) ===')
print(data['Segment'].value_counts())
print('\n')

print('=== DISTRIBUSI KATEGORI PRODUK (Category) ===')
print(data['Category'].value_counts())
```

```python
rata_rata_sales = data['Sales'].mean()
print(f'Rata-rata Penjualan (Sales): {rata_rata_sales:,.2f}')
```

```python
sales_min    = data['Sales'].min()
sales_max    = data['Sales'].max()
sales_median = data['Sales'].median()
sales_std    = data['Sales'].std()
sales_q1     = data['Sales'].quantile(0.25)
sales_q3     = data['Sales'].quantile(0.75)

print('=== STATISTIK PENJUALAN ===')
print(f'  a. Penjualan Minimum   : {sales_min:,.2f}')
print(f'  b. Penjualan Maksimum  : {sales_max:,.2f}')
print(f'  c. Median Penjualan    : {sales_median:,.2f}')
print(f'  d. Standar Deviasi     : {sales_std:,.2f}')
print(f'  e. Kuartil Pertama (Q1): {sales_q1:,.2f}')
print(f'  f. Kuartil Ketiga  (Q3): {sales_q3:,.2f}')
```

```python
print('=== TOP 10 ORDER ID – PENJUALAN TERBESAR ===')
print("\n Top 10 Order ID dengan Penjualan Terbesar :")
print(data.sort_values(by='Sales', ascending=False)[['Order ID','Sales']].head(10))
```

```python
print("\n Top 10 Order ID dengan penjualan terkecil :")
print(data.sort_values(by='Sales', ascending=True)
[['Order ID','Sales']].head(10))
```

```python
kolom_kategori = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category', 'Order Priority']
print("\n Jumlah Transaksi Berdasarkan Category :")
for kolom in kolom_kategori:
 print(f"\n--- {kolom} ---")
 print(data[kolom].value_counts())
```

```python
print("\n Total Penjualan Berdasarkan Category :")
for kolom in kolom_kategori:
    print(f"\n--- {kolom} ---")
    print(data.groupby(kolom)['Sales'].sum())
```

```python
kolom_numerik = ['Quantity', 'Discount', 'Profit', 'Shipping Cost']

print('=== STATISTIK KOLOM NUMERIK ===')
for kol in kolom_numerik:
    print(f'\n  Kolom: {kol}')
    print(f'    Total    : {data[kol].sum():,.2f}')
    print(f'    Rata-rata: {data[kol].mean():,.2f}')
    print(f'    Minimum  : {data[kol].min():,.2f}')
    print(f'    Maksimum : {data[kol].max():,.2f}')
```

```python
print("""\n Interpretasi Singkat Hasil Analisis Penjualan Berdasarkan
Lokasi :""")
print("""Terdapat perbedaan signifikan antar wilayah. Wilayah tertentu
menjadi pasar utama, wilayah lain masih berpotensi dikembangkan.""")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8')
#1. 5 Kota terbesar dengan total penjualan terbesar berdasarkan kolom
print("\n 5 Kota dengan total penjualan terbesar berdasarkan kolom city :")
kota_terbesar = (
    data.groupby('City')['Sales'].sum().sort_values(ascending=False).head(5)
)
print(kota_terbesar)
```

```python
print("\n Visualisasi 5 Kota dengan total penjualan terbesar menggunakan matplotlib atau seaborn:")
plt.figure(figsize=(10,5))
sns.barplot(x=kota_terbesar.values, y=kota_terbesar.index, palette='viridis')
plt.title('5 Kota dengan Total Penjualan Terbesar')
plt.xlabel('Total Penjualan')
plt.show()
```

```python
print("\n 5 State/provinsi dengan total penjualan terbesar berdasarkan kolom state :")
state_terbesar = (
    data.groupby('State')['Sales'].sum().sort_values(ascending=False).head(
        5)
)
print(state_terbesar)
```

```python
print("""\n Visualisasi 5 state/provinsi dengan total penjualan
terbesar menggunakan matplotlib atau seaborn :""")
plt.figure(figsize=(10,5))
sns.barplot(x=state_terbesar.values, y=state_terbesar.index,
palette='plasma')
plt.title('Visualisasi 5 state/provinsi dengan total penjualan terbesar\nmenggunakan matplotlib atau seaborn')
plt.xlabel('Total Penjualan')
plt.show()
```

```python
print("\n 5 Region penjualan terbesar :")
region_terbesar = (
    data.groupby('Region')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)
print(region_terbesar)
```

```python
print("\n Visualisasi total Penjualan berdasarkan Region :")
plt.figure(figsize=(8,4))
sns.barplot(x=region_terbesar.values, y=region_terbesar.index,
palette='magma')
plt.title('Total Penjualan Berdasarkan Region')
plt.xlabel('Total Penjualan')
plt.show()
```

```python
print("\n 5 Sub-Category Dengan Penjualan Terbesar :")
subcat_top = data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(5)
print(subcat_top)
```

```python
print("""\n Visualisasi 5 Sub-Category Barang Dengan Total Penjualan
Terbesar :""")
plt.figure(figsize=(10,5))
sns.barplot(x=subcat_top.values, y=subcat_top.index,
palette='Spectral')
plt.title('5 Sub-Category Barang Dengan Total Penjualan Terbesar')
plt.xlabel('Total Penjualan')
plt.show()
```

```python
print("\n Rata-rata Penjualan Berdasarkan Category :")
print(data.groupby('Category')['Sales'].mean())
```

```python
print('=== RATA-RATA PENJUALAN BERDASARKAN SUB-CATEGORY ===')
subcat_avg = data.groupby('Sub-Category')['Sales'].mean().sort_values(ascending=False).reset_index()
subcat_avg.columns = ['Sub-Category', 'Rata-rata Sales']
subcat_avg['Rata-rata Sales'] = subcat_avg['Rata-rata Sales'].map('{:,.2f}'.format)
print(subcat_avg.to_string(index=False))
```

```python
print('=== TOTAL QUANTITY BERDASARKAN CATEGORY ===')
cat_qty = data.groupby('Category')['Quantity'].sum().sort_values(ascending=False).reset_index()
cat_qty.columns = ['Category', 'Total Quantity']
print(cat_qty.to_string(index=False))
```

```python
print('=== TOTAL QUANTITY BERDASARKAN SUB-CATEGORY ===')
subcat_qty = data.groupby('Sub-Category')['Quantity'].sum().sort_values(ascending=False).reset_index()
subcat_qty.columns = ['Sub-Category', 'Total Quantity']
print(subcat_qty.to_string(index=False))
```

```python
print('=== TOTAL PROFIT BERDASARKAN CATEGORY ===')
cat_profit = data.groupby('Category')['Profit'].sum().sort_values(ascending=False).reset_index()
cat_profit.columns = ['Category', 'Total Profit']
cat_profit['Total Profit'] = cat_profit['Total Profit'].map('{:,.2f}'.format)
print(cat_profit.to_string(index=False))
```

```python
print('=== TOTAL PROFIT BERDASARKAN SUB-CATEGORY ===')
subcat_profit = data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).reset_index()
subcat_profit.columns = ['Sub-Category', 'Total Profit']
subcat_profit['Total Profit'] = subcat_profit['Total Profit'].map('{:,.2f}'.format)
print(subcat_profit.to_string(index=False))
```

```python
def buat_grafik(nama_kolom, judul):
    hasil = (data.groupby(nama_kolom)['Sales']
        .sum()
        .sort_values(ascending=False)
    )
    plt.figure(figsize=(8,4))
    sns.barplot(x=hasil.values, y=hasil.index, palette='Set2')
    plt.title(judul)
    plt.xlabel('Total Penjualan')
    plt.show()
    return hasil
#1 Visualisasi Total Penjualan Berdasarkan Ship Mode
print("\n Visualisasi Penjualan Berdasarkan Ship Mode :")
print(buat_grafik('Ship Mode', """Total Penjualan Berdasarkan Ship
Mode"""))
```

```python
print("\n Total Penjualan berdasarkan Segment :")
print(buat_grafik('Segment', 'Total Penjualan Berdasarkan Segmen'))
```

```python
print("\n Total Penjualan Berdasarkan Region :")
print(buat_grafik('Region', 'Penjualan Berdasarkan region'))
```

```python
print("\n Penjualan berdasarkan Category :")
print(buat_grafik('Category', 'Penjualan Berdasarkan category'))
```

```python
print("\n Penjualan berdasarkan Order Priority :")
print(buat_grafik('Order Priority', """Penjualan Berdasarkan Order
Priority"""))
```

```python
print("\n Kombinasi Region & Category :")
kombinasi1 = data.pivot_table(index='Region', columns='Category',
values='Sales', aggfunc='sum', fill_value=0)
print(kombinasi1)
kombinasi1.plot(kind='bar', stacked=True, figsize=(10,6),
colormap='Set3')
plt.title('Total Penjualan: Kombinasi Region Dan Category')
plt.ylabel('Total Penjualan')
plt.show()
```

```python
print("""\n Total Penjualan Berdasarkan Kombinasi Ship Mode & Category
:""")
kombinasi2 = data.pivot_table(index='Ship Mode', columns='Category',
values='Sales', aggfunc='sum', fill_value=0)
print(kombinasi2)
kombinasi2.plot(kind='bar', stacked=True, figsize=(10,6),
colormap='Set1')
plt.title('Penjualan: Cara Kirim & Kategori')
plt.ylabel('Total Penjualan')
plt.show()
```

```python
print("\n Kombinasi Segment & Category :")
kombinasi3 = data.pivot_table(index='Segment', columns='Category',
values='Sales', aggfunc='sum', fill_value=0)
print(kombinasi3)
kombinasi3.plot(kind='bar', stacked=True, figsize=(10,6),
colormap='Paired')
plt.title('Penjualan: Segment & Category')
plt.ylabel('Total Penjualan')
plt.show()
```

```python
print("\n Profit berdasarkan Order Priority :")
profit_prio = data.groupby('Order Priority')['Profit'].sum()
print(profit_prio)
plt.figure(figsize=(8,4))
sns.barplot(x=profit_prio.index, y=profit_prio.values,
palette='coolwarm')
plt.title('Total Keuntungan Berdasarkan Prioritas Pesanan')
plt.ylabel('Total Profit')
plt.show()
```
